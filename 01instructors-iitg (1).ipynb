```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import lightgbm as lgb
from sklearn.metrics import accuracy_score, classification_report
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
import joblib


try:
    df = pd.read_csv('/kaggle/input/train-data/Dev_data_to_be_shared.csv')
except FileNotFoundError:
    raise FileNotFoundError("Dataset not found. Check the file path.")

df.head(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

corr = df.corr()
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(df.drop('bad_flag', axis=1), df['bad_flag'])

importances = model.feature_importances_
features = df.columns[:-1]
```

```python
numerical_features = df.select_dtypes(include=['float64', 'int64']).columns
if len(numerical_features) == 0:
    raise ValueError("No numeric columns found in the dataset.")

# Checking for columns with missing values
missing_columns = numerical_features[df[numerical_features].isnull().any()]
print(f"Columns with missing values before imputation: {missing_columns}")
```

```python
imputer = SimpleImputer(strategy='mean')
df_imputed_array = imputer.fit_transform(df[missing_columns])
```

```python
retained_columns = missing_columns[:df_imputed_array.shape[1]]


dropped_columns = set(missing_columns) - set(retained_columns)
print(f"Columns dropped during imputation: {dropped_columns}")


df_imputed = pd.DataFrame(df_imputed_array, columns=retained_columns)

for col in dropped_columns:
    df_imputed[col] = 0

df_imputed = df_imputed[missing_columns]

df.update(df_imputed)
```

```python
df_cleaned = df.drop_duplicates()
if 'bad_flag' not in df_cleaned.columns:
    raise KeyError("Target column 'bad_flag' not found in the dataset.")


y = df_cleaned['bad_flag']
if y.nunique() <= 1:
    raise ValueError("Target variable 'bad_flag' must have at least two classes.")


X = df_cleaned.drop(columns=['bad_flag'], errors='ignore')
X_numerical = X.select_dtypes(include=['float64', 'int64'])
```

```python
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X_numerical), columns=X_numerical.columns)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
print("Model started training")
model = LogisticRegression(max_iter=10000)
model.fit(X_train_resampled, y_train_resampled)

predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
report = classification_report(y_test, predictions)

print(f"Accuracy: {accuracy}")
print(f"Classification Report: \n{report}")
```

```python
from sklearn.linear_model import SGDClassifier

svm_model = SGDClassifier(loss='hinge', max_iter=1000, random_state=42)
svm_model.fit(X_train_resampled, y_train_resampled)


predictions = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
report = classification_report(y_test, predictions)

print(f"Accuracy: {accuracy}")
print(f"Classification Report: \n{report}")
```

```python
import tensorflow as tf

scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X_numerical), columns=X_numerical.columns)

# Reshape data for LSTM (BiLSTM expects a 3D input shape: [samples, timesteps, features])
X_scaled_3d = X_scaled.values.reshape((X_scaled.shape[0], X_scaled.shape[1], 1))

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled_3d, y, test_size=0.2, random_state=42)

# Build the BiLSTM Model
model = tf.keras.Sequential([
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64, return_sequences=False), input_shape=(X_train.shape[1], X_train.shape[2])),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')  # For binary classification (use 'softmax' for multi-class)
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
print("Training model using BiLSTM")
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
```

```python
import tensorflow as tf

scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X_numerical), columns=X_numerical.columns)


X_scaled_3d = X_scaled.values.reshape((X_scaled.shape[0], X_scaled.shape[1], 1, 1))

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled_3d, y, test_size=0.2, random_state=42)


model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(64, (3, 1), activation="relu", input_shape=(X_train.shape[1], 1, 1)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Reshape((-1, 64)),  # adapts to Conv2D output
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64, activation="tanh")),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])


# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
print("Training model using BiLSTM")
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
```

```python
predictions = (model.predict(X_test) > 0.5).astype(int)
accuracy = accuracy_score(y_test, predictions)
report = classification_report(y_test, predictions)

print(f"Accuracy: {accuracy}")
print(f"Classification Report: \n{report}")
```
